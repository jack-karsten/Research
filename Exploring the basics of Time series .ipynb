```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import yfinance as yf
from yf_ticker_data import get_ticker_data #pre built in my git
```

The concept of Moving Averages (MA) is fundamentally a method of time series analysis. This is because financial markets are inherently time-ordered sequences, fitting the definition of a time series. By applying a Moving Average, such as an X-day MA, we can effectively smooth our dataset, aiding in the identification and recognition of trends within the market data. The primary objective of this technique is to gain a clearer understanding of potential market trends, which might not be immediately apparent in the raw, day-to-day fluctuations of market prices. This approach allows for a more nuanced analysis of market behavior over time, providing valuable insights for forecasting future trends.

```python
# Using a pre built f to load in data using function for YF data
data = get_ticker_data('NQ=F','max')
```

To calculate a Moving Average (MA) for a specific number of days, we employ the 'rolling' function in data analysis. Then we just mean out the data to create the final MA

```python
data['MA_200'] = data.Close.rolling(window=200).mean() # Setting new column for 200 MA 
data['MA_100'] = data.Close.rolling(window=100).mean() # Setting new column for 100 MA 
data['MA_50'] = data.Close.rolling(window=50).mean() # Setting new column for 50 MA 
data['MA_20'] = data.Close.rolling(window=20).mean() # Setting new column for 20 MA 
```

```python
# Your plotting code
plt.plot(data.iloc[4869:].MA_200.dropna(), label='MA 200')
plt.plot(data.iloc[4869:].MA_100.dropna(), label='MA 100')
plt.plot(data.MA_50.iloc[4869:].dropna(), label='MA 50')
plt.plot(data.MA_20.iloc[4869:].dropna(), label='MA 20')
plt.plot(data.iloc[4869:].Close, label='Close')

# Adding the legend
plt.legend()

# Display the plot
plt.show()
```

In my analysis, which focuses on the most recent 1000 days for a more zoomed-in view of market activity, a closer examination reveals distinct behavioral patterns among the moving averages (MAs). These patterns indicate that the shorter MAs generally reintroduce noise into the equation, contrary to the objective of simplifying market analysis. When examining just the 100-day and 200-day MAs, additional insights emerge. Notably, when they 'cross' each other, it can be observed that the market tends to follow the direction indicated by the 100-day MA 'cross' into the future. This is one way in which time series analysis can provide potentially valuable insights into how the market transitions from one paradigm to the next.
